```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
# Load Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'Iris-setosa',1:'Iris-versicolor',2:'Iris-virginica'})
df.head()
```

```python
# Group by species and calculate statistics
grouped = df.groupby('species')

summary = grouped.agg(['mean','median','min','max','std'])
summary
```

```python
# Separate statistics for each species
setosa = df[df['species']=='Iris-setosa']
versicolor = df[df['species']=='Iris-versicolor']
virginica = df[df['species']=='Iris-virginica']

print("Setosa Mean:\n", setosa.mean(numeric_only=True))
print("\nVersicolor Mean:\n", versicolor.mean(numeric_only=True))
print("\nVirginica Mean:\n", virginica.mean(numeric_only=True))
```

```python
# Percentiles
df.quantile([0.25, 0.5, 0.75], numeric_only=True)
```

```python
# Mode
df.mode()
```

```python
# Range = max - min
range_values = df.max(numeric_only=True) - df.min(numeric_only=True)
range_values
```

```python
# Convert categorical to numerical (One-Hot Encoding)
encoded = pd.get_dummies(df['species'])
encoded.head()
```

```python
# Combine encoded data
df_encoded = pd.concat([df, encoded], axis=1)
df_encoded.head()
```
