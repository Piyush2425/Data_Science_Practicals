## Step 2 — Dataset Description & Source

**Dataset:** Iris Flower Dataset  
**Source:** Kaggle — https://www.kaggle.com/datasets/uciml/iris  
**Original Source:** UCI Machine Learning Repository  

**Description:**  
The Iris dataset was introduced by British statistician R.A. Fisher in his 1936 paper *"The Use of Multiple Measurements in Taxonomic Problems"*. It contains measurements of 150 iris flowers from 3 different species:
- *Iris-setosa*
- *Iris-versicolor*
- *Iris-virginica*

Each species has **50 samples**. The dataset has **6 columns**:

| Column | Type | Description |
|--------|------|-------------|
| Id | Integer | Unique identifier |
| SepalLengthCm | Float | Length of sepal in cm |
| SepalWidthCm | Float | Width of sepal in cm |
| PetalLengthCm | Float | Length of petal in cm |
| PetalWidthCm | Float | Width of petal in cm |
| Species | String (Categorical) | Iris species name |

## Step 5 — Data Formatting & Data Normalization

### 5b. Min-Max Normalization (Data Normalization)

```python
# Min-Max Normalization on SepalLengthCm column
x = df[['SepalLengthCm']].values.astype(float)

min_max_scaler = MinMaxScaler()
x_scaled       = min_max_scaler.fit_transform(x)
df_normalized  = pd.DataFrame(x_scaled, columns=['SepalLengthCm_normalized'])

print("--- Normalized SepalLengthCm (Min-Max Scaling) ---")
df_normalized.head(10)
```

```python
# Normalize ALL numeric columns (excluding Id)
feature_cols   = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
scaler         = MinMaxScaler()
df_all_norm    = df.copy()
df_all_norm[feature_cols] = scaler.fit_transform(df[feature_cols])

print("--- All Numeric Features Normalized ---")
df_all_norm[feature_cols].head(10)
```

## Step 6 — Turning Categorical Variables into Quantitative Variables

### 6a. Label Encoding

```python
# Unique species before encoding
print("Unique values before encoding:", df['Species'].unique())

# Apply Label Encoding
df_label = df.copy()
label_encoder = LabelEncoder()
df_label['Species'] = label_encoder.fit_transform(df_label['Species'])

# Unique species after encoding
print("Unique values after encoding :", df_label['Species'].unique())
print("\nEncoding Map:")
for i, cls in enumerate(label_encoder.classes_):
    print(f"  {cls} → {i}")

print("\n--- Label Encoded DataFrame (first 10 rows) ---")
df_label.head(10)
```

### 6b. One-Hot Encoding (using sklearn)

```python
# Apply One-Hot Encoding
df_ohe = df.copy()

# First apply label encoding to Species
df_ohe['Species'] = label_encoder.fit_transform(df_ohe['Species'])

# Remove Species column for features
features_df = df_ohe.drop(columns=['Species'])

# One-hot encode the Species column
enc       = OneHotEncoder(sparse_output=False)
enc_array = enc.fit_transform(df_ohe[['Species']])
enc_df    = pd.DataFrame(enc_array)

# Join encoded values back to features
df_encode = features_df.join(enc_df)

# Rename the new columns
df_encode.rename(columns={0.0: 'Iris-Setosa', 1.0: 'Iris-Versicolor', 2.0: 'Iris-Virginica'}, inplace=True)

print("--- One-Hot Encoded DataFrame ---")
df_encode.head()
```

### 6c. Dummy Variable Encoding (using pandas get_dummies)

```python
# Dummy encoding with pandas get_dummies (k-1 variables, drop_first=True)
df_dummy = df.copy()
df_dummy = pd.get_dummies(df_dummy, columns=['Species'], drop_first=True)

print("--- Dummy Encoded DataFrame (k-1 dummy variables) ---")
df_dummy.head()
```

```python
# Full one-hot encoding using get_dummies (k dummy variables)
df_dummies_full = pd.get_dummies(df, columns=['Species'], drop_first=False)

print("--- Full One-Hot Encoding via get_dummies ---")
df_dummies_full.head()
```
