# DS Lab 7: Text Analytics using NLTK

## Aim
To perform text preprocessing and feature extraction on a sample paragraph about Cristiano Ronaldo using NLTK and scikit-learn.

## Dataset
A meaningful paragraph about Cristiano Ronaldo.

## Step 1: Import Required Libraries
This step imports all libraries needed for tokenization, stopword removal, stemming, lemmatization, POS tagging, Bag of Words, and TF-IDF.

```python
# Import NLTK tools for preprocessing
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

# Import scikit-learn tools for feature extraction
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

## Step 2: Download NLTK Resources
This step downloads all required corpora and models used by tokenizers, stopwords, lemmatizer, and POS tagger.

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

## Step 3: Define Input Text
A sample paragraph about Cristiano Ronaldo is used as the input document for processing.

```python
text = """Cristiano Ronaldo is one of the greatest football players in the history of the sport.
He was born in Portugal and started his professional career at a young age. 
Ronaldo became famous while playing for Manchester United, where he developed into a world-class player.
Later, he joined Real Madrid and became the club's all-time top scorer, winning multiple 
UEFA Champions League titles. Ronaldo is known for his incredible speed, strength, discipline,
and goal-scoring ability. He has also played for Juventus and returned to Manchester United before continuing 
his journey in professional football. Throughout his career, he has won several Ballon d'Or awards, 
which are given to the best player in the world. Apart from his football skills, Ronaldo is admired for his dedication,
hard work, and leadership qualities. He inspires millions of fans around the world and is also involved in charity work. His journey from 
a young boy in Madeira to a global superstar shows his determination and passion for success."""

print('Input Text:')
print(text)
```

## Step 4: Sentence Tokenization
Sentence tokenization splits text into individual sentences using `sent_tokenize`.

```python
# Split paragraph into sentences
sentences = sent_tokenize(text)

print('Sentences:')
for index, sentence in enumerate(sentences, start=1):
    print(f'{index}. {sentence}')
```

## Step 5: Word Tokenization
Tokenization splits text into words using `word_tokenize`.

```python
# Split paragraph into word-level tokens
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

## Step 6: Stopword Removal
Stopword removal filters common words such as 'the', 'is', and 'and' to retain meaningful terms.

```python
# Remove stopwords and punctuation-like tokens
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word.isalpha() and word.lower() not in stop_words]

print('Filtered Words (After Stopword Removal):\n', filtered_words)
```

## Step 7: Stemming
Stemming reduces words to their root form using `PorterStemmer`.

```python
# Convert words to their stem form
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("Stemmed Words:\n", stemmed_words)
```

## Step 8: Lemmatization
Lemmatization converts words to their dictionary base form using `WordNetLemmatizer`.

```python
# Convert words to their lemma (base) form
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Lemmatized Words:\n", lemmatized_words)
```

## Step 9: POS Tagging
Part-of-speech tagging assigns grammatical labels (noun, verb, adjective, etc.) to each token.

```python
# Tag filtered words with POS labels
pos_tags = pos_tag(filtered_words)

print("POS Tags:\n", pos_tags)
```

## Step 10: Bag of Words (BoW)
Bag of Words represents text by counting the occurrence of each term in each sentence.

```python
# Build Bag of Words representation
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(sentences)

print('BoW Feature Names:\n', bow_vectorizer.get_feature_names_out())
print('\nBoW Matrix:\n', bow_matrix.toarray())
```

## Step 11: TF-IDF
TF-IDF represents the importance of words by balancing frequency in a sentence against frequency across all sentences.

```python
# Build TF-IDF representation
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

print('TF-IDF Feature Names:\n', tfidf_vectorizer.get_feature_names_out())
print('\nTF-IDF Matrix:\n', tfidf_matrix.toarray())
```
